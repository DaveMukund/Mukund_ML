```python
import numpy as np 
x = np.array([1,2,3,4,5])
y = np.array([12,18,25,27,35])
```

```python
n = len(x)
```

```python
sum_x = np.sum(x)
sum_y = np.sum(y)
sum_x2 = np.sum(x**2)
sum_xy = np.sum(x * y)

```

```python
print("sum_x =", sum_x)
print("sum_y =", sum_y)
print("sum_xy =", sum_xy)
print("sum_x2 =", sum_x2)
```

```python
x_mean = np.mean(x)
y_mean = np.mean(y)
x_mean,y_mean
```

```python
m=(n* sum_xy- sum_x*sum_y)/ (n*sum_x2-sum_x**2)
print("\nSlope (m =)",round(m,2))

c = y_mean - m * x_mean
print("intercept (c) =", round(c,2))
print("\nRegression Equation: y ={:.2f} + {:.2f}".format(m,c))
```

```python
y_pred = m*x+c
errors = y- y_pred
print("\nPredicted Values and Errors:")
print("x y_actual y_pred error")
for i in range(n):
    print(x[i],"",y[i],"",round(y_pred[i],2),"",round(errors[i],2))
    
```

```python
SSE = np.sum(errors**2)
MSE = SSE/n
MSE
```

```python
MAE = np.mean(np.abs(errors))
print("MAE =",round(MAE,2))
SST = np.sum((y-y_mean)**2)
MAE,SST
```

```python
R2 = 1-(SSE/SST)
print("\n SST =", round(SST,2))
print("R^2=",round(R2,4))
```

```python
def predict(x_new):
    return m*x_new + c 

print("\n---Prediction for new Input---")
x_input = float(input("enter study hours:"))
y_output = predict(x_input)

print(f"predicted Marks = {y_output:.2f}")
```
